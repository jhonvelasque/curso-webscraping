### Extrayendo información

```python
import requests
from bs4 import BeautifulSoup
```

```python
link='https://www.pagina12.com.ar'
pagina=requests.get(link)
pagina.status_code
```

```python
#parseamos
s = BeautifulSoup(pagina.text, 'lxml')
```

```python
type(s)
```

```python
secciones = s.find('div', attrs={'class':'p12-dropdown-column'}).find_all('a')
print (secciones)
```

```python
primer_seccion= secciones[0]
primer_seccion
```

```python
#link de la seccion
primer_seccion.get('href')
#nombre de la seccion
primer_seccion.get_text()
```

```python
#lista de links de cada seccion
links_secciones = [i.get('href') for i in secciones]
#saber typo de variable
type(links_secciones)
#saber longitud de lista creada de href de secciones
len(links_secciones)
links_secciones[0]
```

```python
primer_link_secciones = requests.get(links_secciones[0])
soup_primer_link_secciones = BeautifulSoup(primer_link_secciones.text, 'lxml')
print(soup_primer_link_secciones.prettify())
```

```python
top_contenido=soup_primer_link_secciones.find('section', attrs={'class':'top-content'})
```

```python
no_top_contenido= soup_primer_link_secciones.findAll('article', attrs={'class':'article-item'})
no_top_contenido
```

```python
links_no_top_contenido = [(link+i.a.get('href')) for i in no_top_contenido]
links_no_top_contenido
```
